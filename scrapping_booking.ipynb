##### Generalization on all hotels available on booking 
##### (case of Morocco and 7 hotels for each city, delete the cm variables (counter) to get all hotels)

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd


li1=[] #liste des villes disponibles avec le nombre des hotels
li2=[] #liste des noms des hotels
li3=[] #liste contenant tous les informations sur les clients ayant essayer un tel hotel
NoneType = type(None)
l_page = requests.get("https://www.booking.com/booking-home/country/ma.fr.html")
l_soup = BeautifulSoup(l_page.content, 'html.parser')
l_soup.prettify()

for i in l_soup.find('ul',{'class':'bui-carousel__inner'}).find_all('a',{'class':'bui-card bui-card--media bui-card--transparent'}):
    link = 'https://www.booking.com'+i.get('href')
    city_name = i.find('div',{'class' : 'bui-card__content'}).find('h3',{'class':'bui-card__title'}).get_text()
    total_hotels = int(i.find('div',{'class' : 'bui-card__content'}).find('h4',{'class':'bui-card__subtitle'}).get_text().replace('\n','').replace(' locations de vacances',''))
    d = [city_name,total_hotels]
    li1.append(d)

    hs_page = requests.get(link)
    hs_soup = BeautifulSoup(hs_page.content, 'html.parser')
    hs_soup.prettify()
    cm = 0

    for k in hs_soup.find_all('header',{'class':'bui-spacer--medium'}):
        hs_link = 'https://www.booking.com'+k.find('a').get('href') 
        hotel_name = hs_link.replace('https://www.booking.com/hotel/ma/','').replace('.fr.html','')
        li2.append(hotel_name)
        cm+=1
        if cm == 7 : 
            break


        h_page = requests.get(hs_link)
        h_soup = BeautifulSoup(h_page.content, 'html.parser')
        h_soup.prettify()
        
        
        for m in h_soup.find_all('li',{'class':'ebb6d69bfc'}):
            if type(m.find('span',{'class':'a5499473ab f01b03907b'})) == NoneType or type(m.find('span',{'class':'a5499473ab f01b03907b'}))==NoneType or type(m.find('div',{'class':'db29ecfbe2 c688f151a2'}))==NoneType:
                c_name=''
                c_country=''
                c_commentary=''
            else :
                c_name = m.find('div',{'class':'f9afbb0024 f0d4d6a2f5'}).get_text()
                c_country = m.find('span',{'class':'a5499473ab f01b03907b'}).get_text()
                c_commentary = m.find('div',{'class':'db29ecfbe2 c688f151a2'}).get_text()
            li3.append([city_name,hotel_name,c_name,c_country,c_commentary])
```
